## Exploratory Data Analysis (EDA)

---


```python
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/Ale-Man1/Diamonds-Analysis/main/diamonds_1.csv') # reading csv file that I modified
df.index += 1 # start index as one
print(df, "\n")
```

The majority of the diamonds seem to be **under one carat**, and the higher amount of carats a diamond has, the less diamonds appear in the set. From the data collected it appears that diamonds with higher carat amount are rare compared to the ones with less carat amount.

## Inferences

#### Inference #1:


The approach to discovering the proportion of diamonds between .30 carat and 1.08 carat was to take the amount of diamonds that have their number of carats between those two (inclusive) and divide it by the amount of diamonds overall. Once that arithmetic operation was completed the percentage that was given was **0.7468298109010011** percent.

```python
# found the way to do it this way from https://stackoverflow.com/questions/31617845/how-to-select-rows-in-a-dataframe-between-two-values-in-python-pandas
infr_1 = len(df[(df['carat']>=0.30 ) & (df['carat']<=1.08)])
infr_1_t = len(df)
infr_1_p = infr_1/infr_1_t
print(infr_1_p)
```

#### Inference #2:

The steps to find all the diamonds with the same x and y values was to find them in a similar fashion to inference #1, find the length of the items in the data set that match "df['x'] == df['y']" and the number that is given from the line of code is **17** diamonds with the same x and y value.

```python
# similar way to the code block above this one
infr_2 = len(df[(df['x'] == df['y'])])
print(infr_2)
```

#### Inference #3:


The way to find how many of the diamonds have a carat value that is less than the mean carat value was to first of all find the mean and find the amount of diamonds with a carat value less than the mean. The amount of diamonds that have a carat value that is less than the mean carat value is **31127**.

```python
mean_3 = df['carat'].mean() # similar way to the code block in inference 1
infr_3 = len(df[(df['carat'] < mean_3)])
print(infr_3)
```

#### Inference #4: 

The way to find how many diamonds have a premium cut or better was to find diamonds that are premium cut and diamonds that are ideal cut since those two are the only onces that are premium or better. When that is found, the number that is given is **35342** diamonds that have a premium cut or better.

```python
infr_4 = len(df[(df["cut"] == "Premium") | (df["cut"] == "Ideal")])
print(infr_4)
```

#### Inference #5: 

One way to discover which diamond has the highest price per carat and what is its value is by dividing each diamonds price by its carat value and collecting the one with the highest value. Once that is done you use the equal function for dataframes and get **diamond number: 27636** and **price per carat: 17828.846154**

```python
infr_5 = df["price"]/df["carat"]
infr_5_1 = infr_5.max()
infr_5_2 = infr_5[infr_5.eq(infr_5_1)]
print(infr_5_2)
```

## Conclusion

One observation from the inferences made on the given data is that the price of a diamond increases the more carats it has. Not only is there a relationship between carats and price, but the majority of diamonds in the given data set are between 0.30 and 1.08 in regards to how many carats they are. Another observation is that premium cut diamonds are the most expensive out of all the cuts but are not the rarest, meaning that a diamond being the most expensive doesn't automatically mean it possesses all the rarest attributes as some diamonds with a higher carat value cost less than some with a lower carat value.
